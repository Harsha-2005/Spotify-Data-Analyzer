```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

```python
df=pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
df['duration_min']=df['duration_ms']/60000
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['popularity'],bins=25,kde=True,color='#1DB954')
plt.title('Distribution of Track Popularity',fontsize=16)
plt.xlabel("Popularity Score (0-100)",fontsize=14)
plt.ylabel("Number of Tracks",fontsize=14)
plt.grid(axis='y',alpha=0.5)
plt.show()
```

```python
feature_cols=['popularity','danceability','energy','loudness','acousticness','instrumentalness','liveness','valence','tempo']
corr_matrix=df[feature_cols].corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt=".2f",linewidth=.5,cbar_kws={'label':'Correlation Coefficient'})
plt.title('Correlation Heatmap of Spotify Audio Features',fontsize=16)
plt.xticks(rotation=45,ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

Top 10 Most Popular Artists (Bar Chart)
Find the artists with the highest average popularity across all their songs in the dataset.

```python
top_artists=df.groupby('artists')['popularity'].mean().nlargest(10)
plt.figure(figsize=(12,7))
sns.barplot(x=top_artists.index,y=top_artists.values,palette='viridis')
plt.title('Top 10 Artists by Average Popularity',fontsize=16)
plt.xlabel('Artist',fontsize=14)
plt.ylabel('Average Popularity Score',fontsize=14)
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each track genre and select the top 10
top_genres = df['track_genre'].value_counts().nlargest(10)

plt.figure(figsize=(10, 6))
# Use a bar plot for frequency
sns.barplot(x=top_genres.index, y=top_genres.values, palette='magma')
plt.title('Top 10 Most Frequent Track Genres', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Number of Tracks', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_10_genres_count.png')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
# Use a regression plot to visualize relationship and trend line. Sample
# a few thousand points for better performance and clarity on a large dataset.
sns.regplot(x='danceability', y='popularity', data=df.sample(5000),
            scatter_kws={'alpha':0.2, 's':10}, line_kws={'color':'red'})
plt.title('Popularity vs. Danceability', fontsize=16)
plt.xlabel('Danceability (0.0 to 1.0)', fontsize=12)
plt.ylabel('Popularity Score', fontsize=12)
plt.savefig('popularity_vs_danceability.png')
plt.show()
```
